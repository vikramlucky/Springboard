## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## 2. Find the top 10 major project themes (using column 'mjtheme_namecode')


```python
df.mjtheme_namecode.value_counts().head(10) #This will not work as we have multiple themes in one project.
```

```python
data = df.mjtheme_namecode
```

```python
theme_namecode = pd.concat([json_normalize(x) for x in data])
name = theme_namecode[theme_namecode.name != ''].drop_duplicates()
name = name.set_index('code')
name
```

## Solution 2 

To check the 10 major project theme, i will count the number of code.

```python
theme_counts = pd.DataFrame(theme_namecode.groupby('code').size(), columns=['counts'])
project_themes = name.join(theme_counts).sort_values('counts', ascending=False)
project_themes.head(10)
```

## Solution 3

```python
code_n = name.name.to_dict() # Series to dict, A result will be index, value (Names)
code_n
```

```python
def fill_in(lst): # fill_in function will take an list with dictionary and will replace value of name with one in code_n dictionary above.
    return [{'code': x['code'], 'name': code_n[x['code']]} for x in lst] # Returns a list, with dictionaries inside.
```

```python
new_df = df.copy() #Making a copy of the original dataframe.
new_df.mjtheme_namecode = new_df.mjtheme_namecode.map(fill_in) #Applying fill_in function to each list using map.
```
